# While Loop

The While Loop is used to execute a set of statements until a given is satisfied

A while loop repeats a block of code as long as a certain condition is true

In While Loop, we use a variable for the increment purpose

Increment to this variable is necessary otherwise the loop will never stop and it will run forever

-----

```python
# Initialization
# Condition
# Body of loop
```

```python
# While Loop example 1

# Initializing a variable
# Giving a condition
# Incrementing the variable by 1

i = 0

while i<5:
    print(i)
    #print(i, end=" ")   # to print in one single line
    i = i + 1
```

```python
# While Loop example 2

i = 0
while i<5:
    i = i + 1
    print(i)
```

```python
# If we don't give increment to the variable
# Then, the loop will run infinite times

i = 0
while i<5:
    print(i)
```

```python
# While Loop example 3

i = 10

while i>=5:
    print(i)
    i = i - 1
    
print("This loop ends here")
```

```python
# While Loop example 4

a = 0
while a<5:
    print(a)
    a=a+1
    print("This loop ends here")
```

```python
# While Loop example 5

x = 0
while x<10:
    print(x)
    x=x+2
```

```python
# While Loop example 6
  
x = 1
while x<10:
    print(x+10)   # The value of x will not change, it will only add 10 to x
    x=x+1
```
